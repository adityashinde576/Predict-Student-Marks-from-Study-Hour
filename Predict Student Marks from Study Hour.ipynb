```python
# -------------------------------
# Linear Regression: Study Hours vs Marks
# -------------------------------

# Import necessary libraries
import pandas as pd                 # For data manipulation
import numpy as np                  # For numerical operations
import matplotlib.pyplot as plt     # For visualization
from sklearn.linear_model import LinearRegression   # Linear regression model
from sklearn.metrics import mean_squared_error, r2_score  # Model evaluation metrics

# -------------------------------
# 1. Prepare the dataset
# -------------------------------
data = {
    "Hours_studied": [1, 2, 3, 4, 5, 6, 7],  # Feature: Hours studied
    "Marks": [35, 40, 50, 60, 70, 75, 85]   # Target: Marks obtained
}

df = pd.DataFrame(data)  # Convert dictionary to DataFrame
print("Dataset:\n", df)

# Separate features (X) and target (y)
X = df["Hours_studied"].values.reshape(-1, 1)  # Reshape to 2D array for sklearn
y = df["Marks"].values

# -------------------------------
# 2. Train Linear Regression model
# -------------------------------
model = LinearRegression()  # Initialize model
model.fit(X, y)              # Train the model on data

# Print model parameters
print("\nModel Training Completed.")
print(f"Slope (m): {model.coef_[0]:.2f}")       # Coefficient (rate of change)
print(f"Intercept (b): {model.intercept_:.2f}") # Intercept (starting value)

# -------------------------------
# 3. Make predictions
# -------------------------------
# Predict marks for specific study hours
pred_5 = model.predict([[5]])
pred_9 = model.predict([[9]])
pred_12 = model.predict([[12]])

# Print predictions
print("\nPredictions:")
print(f"Marks for 5 hours: {pred_5[0]:.2f}")
print(f"Marks for 9 hours: {pred_9[0]:.2f}")
print(f"Marks for 12 hours: {pred_12[0]:.2f}")

# -------------------------------
# 4. Visualize results
# -------------------------------
plt.scatter(df["Hours_studied"], df["Marks"], label="Actual Data", color='blue')   # Scatter plot of actual data
plt.plot(df["Hours_studied"], model.predict(X), label="Regression Line", color='red')  # Regression line
plt.xlabel("Hours Studied")  # X-axis label
plt.ylabel("Marks")          # Y-axis label
plt.title("Linear Regression - Study Hours vs Marks")  # Plot title
plt.legend()                 # Show legend
plt.grid(True)               # Add grid for better readability
plt.show()

# -------------------------------
# 5. Evaluate the model
# -------------------------------
y_pred = model.predict(X)                # Predict on training data
mse = mean_squared_error(y, y_pred)     # Mean Squared Error
r2 = r2_score(y, y_pred)                # R² Score (goodness of fit)

# Print evaluation metrics
print("\nModel Evaluation:")
print(f"MSE (Mean Squared Error): {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```
